```python
%matplotlib inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
img = cv2.imread('LAPTOP.jpg')
cv2_imshow(img)
```

```python
plt.figure(figsize=(5,5))
pic = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(pic)
plt.axis('off')
plt.show()
```

### Translation

```python
# img = cv2.imread('hello.jpg')
print("Translation")
tx,ty=50,100
M = np.float32([[1,0,tx],[0,1,ty]])
translated_img=cv2.warpAffine(img,M,(img.shape[1],img.shape[0]))
plt.figure(figsize=(5,5))
pic = cv2.cvtColor(translated_img, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(pic)
```

### ROTATION

```python
img = cv2.imread('LAPTOP.jpg')
row = img.shape[0]
col = img.shape[1]
theta = 30
center = (col / 2, row / 2)
rot_mat = cv2.getRotationMatrix2D(center, theta, 1)
rotated_img = cv2.warpAffine(img, rot_mat, (col, row))
pic = cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(5, 5))
plt.axis('off')
plt.imshow(pic)
plt.show()
```

### SCALLING



```python
scale_x, scale_y = 10, 10
scaled_img = cv2.resize(img, None, fx=scale_x, fy=scale_y,interpolation=cv2.INTER_LINEAR)
plt.figure(figsize=(5,5))
pic = cv2.cvtColor(scaled_img, cv2.COLOR_BGR2RGB)
plt.imshow(pic)
```

### SHEARING

```python
img=cv2.imread('LAPTOP.jpg')
shear_mat=np.float32([[1,0.2,0],[0.2,1,0]])
sheared_img=cv2.warpAffine(img,shear_mat,(img.shape[1],img.shape[0]))
# plt.figure(figsize=(5,5))
pic = cv2.cvtColor(sheared_img, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(pic)
```

### REFLECTION

```python
img= cv2.imread('LAPTOP.jpg')
reflect_mat=np.float32([[-1,0,img.shape[1]],[0,1,0]])
reflected_img=cv2.warpAffine(img,reflect_mat,(img.shape[1],img.shape[0]))
pic = cv2.cvtColor(reflected_img, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(pic)
```
